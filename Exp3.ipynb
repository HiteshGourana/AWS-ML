WAP to caluclate the sum of even number given by user ?


```python
n=int(input("Enter the size of array "))
print("Enter the number ")
even=0
odd=0
for i in range(n):
    x=int(input())
    if x%2==0:
        even=even+x
    else:
        odd=odd+x

print(f"Odd sum is {even}\nEven Sum is {odd}")
```

Wap to find the greatest of 4 number given by user ?


```python
arr=list()
n=int(input("Enter the number "))
for i in range(n):
    x=int(input())
    arr.append(x)

arr.sort()
print(arr[n-1])
```
