```python
def Function(n):
  if n==1:
    return 1
  return n*Function(n-1)

Function(10)
```

```python
def decorator(func):
  def wrapper():
    print("Hello my name is ",end="")
    print(func())
    print("Good bye")
  return wrapper;

@decorator
def Greeting():
  return "Hitesh"

Greeting()

```

```python
class Person:
    def __init__(self, id, name, age):
        self._id = id
        self.name = name
        self._age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, new_age):
        self._age = new_age
```
