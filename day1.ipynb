```python
def Function(n):
  if n==1:
    return 1
  return n*Function(n-1)

Function(10)
```

```python
def decorator(func):
  def wrapper():
    print("Hello my name is ",end="")
    print(func())
    print("Good bye")
  return wrapper;

@decorator
def Greeting():
  return "Hitesh"

Greeting()

```
